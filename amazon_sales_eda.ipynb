```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

# mengubah tipe data
replace ilakukan 2x karena ada data yang mengandung koma

```python
df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ', '')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype(float)

df['actual_price'] = df['actual_price'].str.replace('‚Çπ', '')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype(float)


df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
```

```python
df['rating'].unique()
```

```python
# mencari tahu data rating | ada di produk mana aja

data = df.loc[df['rating'] == '|'].value_counts()
print(data)
```

```python
#  setelah mengetahui ubah angka itu berdasarkan baris yg sama di product id
df.loc[(df['product_id'] == 'B08L12N5H1') & (df['rating'] == '|'), 'rating'] = '3.9'
df['rating'].unique()
# cara 2
# df.loc[df.product_id == "B08L12N5H1", 'rating'] = "3.9"
```

```python
# mengubah tipe data ke float
df['rating'] = df['rating'].astype(float)
df['rating'].dtype
```

# Handling Missing Value

```python
df.isnull().sum()
```

```python
#  ada 1 kolom memiliki rating count kosong, coba kita cek
df[df['rating_count'].isnull()]
```

```python
# mengubah tipe datanya dulu sebelum diisi dengan nilai median
df['rating_count'] = df['rating_count'].str.replace(',','')
df['rating_count']
df['rating_count'] = df['rating_count'].astype(float)
```

```python
# # mengisi nilai kosong dengan nilai median
df.fillna(value=df['rating_count'].median(), inplace=True)
```

```python
df['rating_count'].isnull().sum()
```

# Overlayed Price

```python
# Menentukan jumlah bin dan rentang bin yang sama dengan Plotly
bins = 10
range_bin = (0, 20000)

# Membuat histogram dengan seaborn
plt.figure(figsize=(10, 6))

# Menggunakan histplot dengan bar stacking
sns.histplot(df['actual_price'], bins=bins, binrange=range_bin, color="#232F3E", label="Actual Price", alpha=0.75)
sns.histplot(df['discounted_price'], bins=bins, binrange=range_bin, color="#FF9900", label="Discounted Price", alpha=0.75)

# Mengatur label dan judul
plt.title("Actual Price x Discounted Price")
plt.xlabel("Price")
plt.ylabel("Number of Products")
plt.legend()

# Menampilkan plot
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Menambahkan garis grid pada sumbu y
plt.tight_layout()
plt.show()
```
