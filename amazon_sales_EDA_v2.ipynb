```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
# Misalnya, jika nilai dalam kolom discount_percentage adalah '25%', ini sebenarnya berarti 25% dari nilai aslinya. Namun,
# ketika Anda ingin menggunakan nilai ini dalam perhitungan matematika (seperti penjumlahan, pengurangan, perkalian),
# Anda perlu mengonversi nilai tersebut menjadi desimal, yaitu 0,25.

df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)
df['discount_percentage'] = df['discount_percentage'] / 100
df['discount_percentage']
```

# Handling Missing Value

## mengubah rating

```python
# elakukan query pada DataFrame df untuk mencari baris di mana nilai kolom rating adalah '|'
df.query('rating == "|"')
# df[df['rating'] == '|'] ini sama aja seperti diatas
# df['rating'] == '|' = kalau ini akan mengembalikan boolean
```

```python
# mengganti | dngan angka 3.9
df['rating'] = df['rating'].str.replace('|', '3.9').astype(float)
```

```python
# merubah tipe data rating kolom
df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

```python
# mencari data kosong dan diurutkan secara descending
df.isnull().sum().sort_values(ascending=False)
```

```python
# menghitung presentasi missing value
round(df.isnull().sum() /len(df)*100,2 ).sort_values(ascending=False)
```

# Rata Rata rating dari kategori poduk

```python
import pandas as pd

# cek tipe data di rating
print(df["rating"].dtype)

# jika tipe data bukan numerik maka ubah tipe datanya
if df["rating"].dtype == "object":
    df["rating"] = pd.to_numeric(df["rating"], errors="coerce")  # Handle potential errors

# Calculate the average ratings after ensuring numeric data type
average_ratings = df.groupby("category")["rating"].mean().reset_index()

print(average_ratings)
```

# Kategori dengan rating tertinggi

```python
# kelompokkan data berdasarkan kategori
average_ratings = df.groupby("category")["rating"].mean().reset_index()

# Sorting
average_ratings = average_ratings.sort_values(by="rating", ascending=False)

# cetak 5 kategori dengan rating tertingi
print("Top 5 categories with highest average ratings:")
for i in range(5):
    category = average_ratings.iloc[i]["category"]
    average_rating = average_ratings.iloc[i]["rating"]
    print(f"{i+1}. {category}: {average_rating:.2f}")
```

```python
# Plot bar chart
plt.figure(figsize=(22,9))
plt.bar(average_ratings["category"].head(), average_ratings["rating"].head(), color='skyblue')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Top 5 Categories with Highest Average Ratings')
plt.xticks(rotation=45)
plt.show()
```
